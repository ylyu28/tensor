```python
import pickle
import tensorly as tl
import numpy as np
import pandas as pd
import tlviz
from sklearn.preprocessing import scale
from tensorly.decomposition import parafac2
from tlviz.utils import normalise
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy.linalg as la
import sspa
import scipy.stats as stats
from pathway_feature_selection import bootstrap_feature_select_pr2, feature_select_bs_ci, significant_pathways, harmonic_pvalue,combined_pseudoz, significant_pseudoz, plot_zscore_distribution
```

```python
with open('p2_common_metab_pseudoz.pkl','rb') as f:
    p2_common_metab_pseudoz = pickle.load(f)

with open('p2_common_prot_pseudoz.pkl','rb') as f:
    p2_common_prot_pseudoz = pickle.load(f)

with open('p1_pseudoz.pkl','rb') as f:
    p1_pseudoz = pickle.load(f)
```

```python
p1_f1_rank = p1_pseudoz.sort_values(by='Factor 1',ascending=False).index
```

```python
p1_f2_rank = p1_pseudoz.sort_values(by='Factor 2',ascending=False).index
```

```python
p2_common_metab_f1 = p2_common_metab_pseudoz.sort_values(by='Factor 1',ascending=False).index
```

```python
p2_common_metab_f2 = p2_common_metab_pseudoz.sort_values(by='Factor 2',ascending=False).index
```

```python
p2_common_prot_f1 = p2_common_prot_pseudoz.sort_values(by='Factor 1',ascending=False).index
```

```python
p2_common_prot_f2 = p2_common_prot_pseudoz.sort_values(by='Factor 2',ascending=False).index
```

```python
jaccard_coefficients = []
indices = range(1, len(p1_f1_rank) + 1)  # From 1 to 60

# Calculate Jaccard coefficients incrementally
for n in indices:
    subset1 = set(p1_f1_rank[:n])
    subset2 = set(p2_common_metab_f1[:n])
    
    intersection = subset1.intersection(subset2)
    union = subset1.union(subset2)
    
    if len(union) > 0:  
        jaccard_coefficient = len(intersection) / len(union)
    else:
        jaccard_coefficient = 0  
    
    jaccard_coefficients.append(jaccard_coefficient)

plt.figure(figsize=(10, 5))
plt.plot(indices, jaccard_coefficients, marker='o', linestyle='-', color='skyblue')
plt.title('Jaccard Coefficient as Lists Expand')
plt.xlabel('Number of Elements')
plt.ylabel('Jaccard Coefficient')
plt.ylim(0, 1)  
plt.grid()
plt.xticks(range(0,61,5))  
plt.show()
```
